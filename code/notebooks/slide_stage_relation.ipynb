```python
import pandas as pd
```

```python
slide_path = "D:\\onedriveudc\\OneDrive - Universidade da Coruña\\tfg\\python\\data\\slide.tsv"
clinical_path = "D:\\onedriveudc\\OneDrive - Universidade da Coruña\\tfg\\python\\data\\clinical.tsv"
manifest = "D:\\onedriveudc\\OneDrive - Universidade da Coruña\\tfg\\python\\data\\manifest.txt"
output = "D:\\onedriveudc\\OneDrive - Universidade da Coruña\\tfg\\python\\data\\slide_stage_relation.csv"
```

```python
#reads slide tsv and clinical tsv
slide = pd.read_csv(slide_path, sep='\t')
clinical = pd.read_csv(clinical_path, sep='\t')
manifest = pd.read_csv(manifest, sep='\t')
```

```python
df = slide[['slide_submitter_id', 'slide_id',"case_id"]].copy()
df["filename"] = df["slide_submitter_id"] + "." + df["slide_id"] + ".svs"
df
```

```python
clinical.drop_duplicates('case_id', inplace = True)
clinical.set_index("case_id", inplace=True)
clinical
```

```python
df2 = df[["case_id","filename","slide_id"]].copy()
#maintain only df2 rows which has the same filename as in manifest
df3 = df2[df2["filename"].isin(manifest["filename"])]
#df2.apply(lambda x: x["case_id"] in x["name"], axis=1)
#maintain only df2 rows which case_id is substring of the manifest id by using apply
#df2 = df2[df2["filename"].apply(lambda x: x in manifest["filename"])]
#df2 = df2[df2["case_id"].isin(manifest["id"])]
#df3 = df2[df2["filename"].str.contains(manifest["filename"])]
df2.set_index("case_id", inplace=True)
df3.set_index("case_id", inplace=True)
df3["ajcc_pathologic_stage"] = clinical["ajcc_pathologic_stage"]
df3
```

```python
df2["ajcc_pathologic_stage"] = clinical["ajcc_pathologic_stage"]
df2
#for row in df2:
#    row["ajcc_pathologic_stage"] = clinical.loc[clinical["case_id"] == row["case_id"]]["ajcc_pathologic_stage"]
```

```python
#all the rows which don`t have -TS followed by a character and a . in the filename are removed
definitive = df2[df2["filename"].str.contains("-TS?.")]
definitive
```

```python
#combine the two dataframes definitive and df3 with different filename
df4 = pd.concat([definitive, df3])
df4.drop_duplicates('filename', inplace = True)
df4.drop(columns=["slide_id"], inplace=True)
df4.reset_index(inplace=True)
df4
```

```python
#simplify the the pathologic state in 4 stages
def simplify(x):
    #check if x is string
    if isinstance(x, str):
        if "Stage IV" in x:
            return 4
        elif "Stage III" in x:
            return 3
        elif "Stage II" in x:
            return 2
        elif "Stage I" in x:
            return 1
        else:
            return 0
    else:
        return 0

df4["stage"] = df4["ajcc_pathologic_stage"].apply(simplify)
#df4["ajcc_pathologic_stage"] = df4["ajcc_pathologic_stage"].str.replace("-1", "I")
df4.to_csv(output, sep='\t', index=False)
df4
```
